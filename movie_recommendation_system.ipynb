```python
# importing the dependencies

import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# data collection

movies_data=pd.read_csv('/content/movies.csv')
movies_data.head()
```

```python
# getting the shape of the data

movies_data.shape
```

```python
# selecting the features form the data

selected_features=['title','cast','crew']
print(selected_features)
```

```python
# combining all the three features

combined_features=movies_data['title']+' '+movies_data['cast']+' '+movies_data['crew']
print(combined_features)
```

```python
# cheking is there any null values

combined_features.isnull().sum()
```

```python
# there is no null values so we can move forward
```

```python
# converting the text data into feature vector

vectorizer=TfidfVectorizer()
feature_vector=vectorizer.fit_transform(combined_features)
print(feature_vector)
```

COSINE SIMILARITY BLOCK

```python
# getting the similarity using the cosine similarty

similarity=cosine_similarity(feature_vector)
print(similarity)
```

```python
# getting the shape of the similart

print(similarity.shape)
```

```python
# getting movie from the user

movie_name=input('enter the movie name: ')
```

```python
# creating a list with all the movie name given in the dataset

list_of_all_titles=movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name

find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
# finding the index of tghe movie of the movie with the name

index_of_the_movie=movies_data[movies_data.title==close_match].index.values[0]
print(index_of_the_movie)
```

```python
# getting the list of similar movie

similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on tghe similarity score

sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)

```

```python
# printing the similar movie based on the index

print("movie suggested for you")

i=1
for movie in sorted_similar_movies:
  index=movie[0]
  titel_from_index=movies_data[movies_data.index==index].title.values[0]
  if(i<30):
    print(i,'.',titel_from_index)
    i+=1
```
